### 1.2 各保存20首李白、杜甫、白居易、王維（維是繁体）的诗（题目以及内容）。要求为每首诗歌生成独立的数据对象，包括 标题、作者、诗内容（按行）的元组（tuple）

```python
import xml.dom.minidom
import xml.sax
DOMTree = xml.dom.minidom.parse("tang300.xml")
collection = DOMTree.documentElement
 
# 在集合中获取所有诗句
poems = collection.getElementsByTagName("作业用唐诗")

class Poem():
    def __init__(self, _t, _a = "", _c = ()):
        self.title = _t
        self.author = _a
        self.content = _c
        
    def __str__(self):
        return f"{self.author}《{self.title}》\n{self.content}"
        
poem_dict = {}
watchlist = ["李白","杜甫","白居易","王維"]
author_dict = dict((author,[]) for author in watchlist)

for line in poems:
    f = lambda x : line.getElementsByTagName(x)[0].childNodes[0].data
    poem_id, line_number, contance = int(f('Poem_id')), int(f('line_number')), f('contance').strip("$#")
    # Title
    if line_number == -100:
        poem_dict[poem_id] = Poem(contance)
    # Author
    elif line_number == -1:
        poem_dict[poem_id].author = contance
        if contance in watchlist:
            author_dict[contance].append(poem_dict[poem_id])
    # Content
    elif 1 <= line_number <= 10:
        poem_dict[poem_id].content += (contance,)

for key,lst in author_dict.items():
    for idx, poem in zip(range(20),lst):
        print(idx+1, poem, "\n")   
```

## 第二部分 发布者-订阅者模式

实现 设计模式中的观察者模式，本部分将以第一部分中提取的诗歌信息作为主题。

观察者模式也被称为发布-订阅（Publish/Subscribe）模式。当这个主题对象状态变化时，会通知所有观察者对象并作出相应处理逻辑。

在本问题中，可以形象地理解成，Publisher为诗歌发布中心，Reader为订阅相应诗人的诗歌的人。需要实现订阅，退订，Publisher可以发布新增、删除诗歌的消息以及当前各位诗人的诗歌的总数。当Publisher发布消息时，它需要把需要推送给所有订阅的Reader。


**Publisher（发布者）**

    - 保存订阅的读者
    - 记录四位诗人各自诗歌的总数
    - 发布四位诗人新诗的消息，每次只需要发布一首
    - 发布新消息时，发布给所有已订阅的读者   
   只有一位发布者Publisher
   
**Reader（订阅者）**

    - 订阅/取消订阅 发布者（注意同一个发布者不要重复订阅）
    - 保存发布者发来的诗歌
关于保存消息，读者分为**两种类型**：
- Type1：保存所有消息，这意味着这类读者需要将四位诗人的诗歌分类。
- Type2：只保存最新的消息（也就是说只需要保存一首诗）。

要求Reader实现printStatistics方法来打印保存的诗歌。

### 2.1 完成上述Publisher & Reader的代码，先不需要考虑多进程

```python
class BasePublisher(object):
    def __init__(self,name):
        self.name = name
        self.readers = []
        self.poems = {}
        
    def subscribeReader(self, reader):
        if reader in self.readers:
            print(f"{reader.name} has already subscribed to you")
        else:
            self.readers.append(reader)

    def unsubscribeReader(self, reader):
        if reader in self.readers:
            self.readers.remove(reader)
        else:
            print(f"{reader.name} hasn't subscribed to you yet!")

    def notifyReader(self, poem, cmd):
        if cmd == "publish":
            for r in self.readers:
                r.receivePoem(self, poem)
        elif cmd == "retract":
            for r in self.readers:
                r.retractPoem(self, poem)
    
    def publish(self, poem):
        author = poem.author
        if author in self.poems:
            self.poems[author].append(poem)
        else:
            self.poems[author] = [poem]
        self.notifyReader(poem, "publish")
    
    def retract(self, poem):
        author = poem.author
        if poem in self.poems[author]:
            self.poems[author].remove(poem)
            self.notifyReader(poem, "retract")
    
    def __str__(self):
        d = dict([(x,len(self.poems[x])) for x in self.poems])
        return f"[{self.name} info] Subscribers:{[x.name for x in self.readers]}, Poems:{d}"

class Publisher(BasePublisher):
    def __init__(self, name):
        BasePublisher.__init__(self, name)
        
```

```python
import threading
class BaseReader(object):
    
    def __init__(self, name):
        self.name = name
        self.subs = []
        
    def subscribeToPublisher(self, publisher):
        publisher.subscribeReader(self)
        if publisher in self.subs:
            print(f"You have already subscribed to {publisher.name}!")
        else:
            self.subs.append(publisher)
            
    def unsubscribeToPublisher(self, publisher):
        publisher.unsubscribeReader(self)
        if publisher in self.subs:
            self.subs.remove(publisher)
        else:
            print(f"You haven't subscribe to {publisher.name} yet!")
        
    def __str__(self):
        return f"{self.name} subs to {[x.name for x in self.subs]}"
    
```

```python
class ReaderType1(BaseReader):

    def __init__(self, _name):
        BaseReader.__init__(self, _name)
        self.magz = {}
        
    def receivePoem(self, publisher, poem):
        print('thread id :{}'.format(threading.current_thread().name))
        author = poem.author
        if author not in self.magz:
            self.magz[author] = [poem]
        else:
            self.magz[author] += [poem]
    
    def retractPoem(self, publisher, poem):
        print('thread id :{}'.format(threading.current_thread().name))
        author = poem.author
        if poem in self.magz[author]:
            self.magz[author].remove(poem)
            
    def printStatistics(self):
        d = dict([(x,len(self.magz[x])) for x in self.magz])
        print(f"{self.name}, Stats:{d}")
        for author,poems in self.magz.items():
            for poem in poems:
                print(poem,"\n")
        print()

# 第二种读者
class ReaderType2(BaseReader):

    def __init__(self, _name):
        BaseReader.__init__(self, _name)
        self.magz = None

    def receivePoem(self, publisher, poem):
        print('thread id :{}'.format(threading.current_thread().name))
        self.magz = poem
        
    def retractPoem(self, publisher, poem):
        print('thread id :{}'.format(threading.current_thread().name))
        if self.magz == poem :
            self.magz = None

    def printStatistics(self):
        print(f"{self.name}\n{self.magz}\n")
```

### 2.2 修改Publisher部分代码，用多线程（或者多进程）的方式发送消息。本部分为选做部分。

```python
class MultiThreadPublisher(BasePublisher):
    def __init__(self,name):
        BasePublisher.__init__(self, name)
        
    def notifyReader(self, poem, cmd):
        th = []
        for r in self.readers:
            f = r.receivePoem if cmd == "publish" else r.retractPoem
            t = threading.Thread(target=f, args=(self,poem,))
            t.start()
            th.append(t)
        for t in th:
            t.join()
```

```python
MyMultiThreadPublisher = MultiThreadPublisher("MultiPublisher")
Alice.subscribeToPublisher(MyMultiThreadPublisher)
Bob.subscribeToPublisher(MyMultiThreadPublisher)
Carol.subscribeToPublisher(MyMultiThreadPublisher)
print(Alice)
```
